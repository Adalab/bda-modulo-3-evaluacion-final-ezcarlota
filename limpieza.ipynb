```python
nulos_clientes =(df_clientes.select_dtypes(include="number").isnull().sum() / df_clientes.shape[0]) * 100

df_nulos_clientes = nulos_clientes.reset_index() #calculamos el porcentaje de nulos
```

```python
df_nulos_clientes  #cambiamos nombre de las columnas de dataframe 

df_nulos_clientes.rename(columns={"index" : "nombre_columna", 0 : "% nulos"}, inplace=True)

df_nulos_clientes
```

```python
mascara_nulos_clientes = df_nulos_clientes["% nulos"] > 0 #filtramos todas las columnas que tengan nulos

df_nulos_clientes = df_nulos_clientes[mascara_nulos_clientes]
df_nulos_clientes
```

```python
# Para imputación de nulos usando métodos estadísticos avanzados
from sklearn.impute import SimpleImputer 
from sklearn.experimental import enable_iterative_imputer # Necesario para IterativeImputer
from sklearn.impute import IterativeImputer 
from sklearn.impute import KNNImputer 



#. Configuración de Pandas para visualizar todas las columnas
pd.set_option('display.max_columns', None) 
```

```python
#como tengo muchos nulos lo hago por tecnicas avanzadas
lista_imputacion_avanzada = ["Salary","Cancellation Year","Cancellation Month"]
```

```python
# 1. Definimos el imputer
knn_imputer = KNNImputer(n_neighbors=5)

# 2. Elegimos las columnas de origen
columnas_originales = ["Salary", "Cancellation Year", "Cancellation Month"]

# 3. Definimos los nombres de las nuevas
columnas_nuevas = ["Salary knn", "Cancellation Year knn", "Cancellation Month knn"]

# 4. Ejecutamos y ASIGNAMOS 
df_clientes[columnas_nuevas] = knn_imputer.fit_transform(df_clientes[columnas_originales])

# 5. Verificamos
df_clientes.head(2)
```

```python
#definimos el imputer

iterative_imputer = IterativeImputer(random_state=42, max_iter=20) 

#elegimos columnas de origen
columnas_originales = ["Salary", "Cancellation Year", "Cancellation Month"]

# 3. Definimos los nombres de las nuevas
columnas_nuevas_iter =["Salary iter","Cancellation Year iter", "Cancellation Month iter"]

# 4. Ejecutamos y ASIGNAMOS 
df_clientes[columnas_nuevas_iter] =iterative_imputer.fit_transform(df_clientes[columnas_originales])

# 5. Verificamos
df_clientes.head(2)

```

```python
df_clientes[["Salary","Salary knn","Salary iter","Cancellation Year","Cancellation Year knn","Cancellation Year iter","Cancellation Month","Cancellation Month knn","Cancellation Month iter"]].agg(["mean","median"]).T
# Miramos a ver cuanto altera al original
```

```python
df_clientes.drop(columns=["Salary","Salary knn","Cancellation Year","Cancellation Year knn","Cancellation Month","Cancellation Month knn"],inplace=True)
```

```python
df_clientes.columns
```

```python
df_clientes.rename(columns={"Salary iter" : "Salary" , "Cancellation Year iter" : "Cancellation Year" , "Cancellation Month iter" : "Cancellation Month"},inplace=True)
```

```python
cols_enteras = ["Cancellation Year", "Cancellation Month"]

# 1. Redondeamos para que 2016.2 sea 2016.0
# 2. Convertimos a Int64 (el que permite nulos y es más robusto)
df_clientes[cols_enteras] = df_clientes[cols_enteras].round(0).astype("Int64")

# Verificamos el resultado
df_clientes.head(2)
```
